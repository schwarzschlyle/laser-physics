# Spherical-Spherical Mirror Resonator

```python
adjustment_factor = 10
R = 4
reflections = 5

for i in range(20):
    L = (0.2*i)*R
    path = LaserPath()
    path.append(Space(d=L/2))
    path.append(CurvedMirror(R=-R*adjustment_factor))
    path.append(Space(d=-L))
    path.append(CurvedMirror(R=R*adjustment_factor))
    #path.display(beams=[GaussianBeam(w=0.001*adjustment_factor)])

    for i in range(reflections):
        path.append(Space(d=L))
        path.append(CurvedMirror(R=-R*adjustment_factor))
        path.append(Space(d=-L))
        path.append(CurvedMirror(R=R*adjustment_factor))    
        #path.display(beams=[GaussianBeam(w=0.001*adjustment_factor)])
    #print("L = " + str(L) + ", R = " + str(R))
    path.display(beams=[GaussianBeam(w=0.001*adjustment_factor)], comments = "L = " + str(L) + ", R = " + str(R))
```

# Spherical-Planar Mirror Resonator

```python
adjustment_factor = 10 # empirical adjustment factor to utilize a physically-accurate simulation
R = 4
reflections = 5

for i in range(20):
    L = (0.2*i)*R # vary this
    path = LaserPath()
    path.append(Space(d=L/2))
    path.append(CurvedMirror(R=-100000))
    path.append(Space(d=-L))
    path.append(CurvedMirror(R=R*adjustment_factor))
        #path.display(beams=[GaussianBeam(w=0.001*adjustment_factor)])

    for i in range(reflections):
        path.append(Space(d=L))
        path.append(CurvedMirror(R=-100000))
        path.append(Space(d=-L))
        path.append(CurvedMirror(R=R*adjustment_factor))    
            #path.display(beams=[GaussianBeam(w=0.001*adjustment_factor)])
    # print("L = " + str(L) + ", R = " + str(R))
    path.display(beams=[GaussianBeam(w=0.001*adjustment_factor)], 
                 comments = "L = " + str(L) + ", R = " + str(R))
```
